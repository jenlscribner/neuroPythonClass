## Part 2: Plotting US and Italy Data

We can extract the data for the US from the dataframe

```python
df_us = df[df['Country/Region'] == 'US']
df_us = df_us.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_us = df_us.drop(columns=['Lat','Long','Province/State'])

data_us = df_us.loc['US','2/20/20':'4/5/20']
```

## 2a: Plot Italy and US data in two axes

Use subplots and assign separate colors to each country. Set the same limits on the y-axis for both subplots

```python
# use 
fig,ax = plt.subplots(1,2,figsize=(15,5)) # or whatever dimensions you like

ax[0].plot(data_italy.values,'.-',color='C0',label='COVID-19 in Italy')
ax[1].plot(data_us.values,'.-',color='C1',label='COVID-19 in US')

for a in ax:
    a.set_ylim(0,350000)
    a.set_xlabel('days')
    a.set_ylabel('# of cases')
    a.legend(fontsize=12)
    
plt.show()
```

## 2b: Plot Italy and US data in the same plot

Use the same colors as above, and include a legend

```python
# your code here
plt.figure(figsize=(12,5))

plt.plot(data_italy.index,data_italy.values,'.-',color='C0',label='COVID-19 in Italy')
plt.plot(data_us.index,data_us.values,'.-',color='C1',label='COVID-19 in US')

plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('# of cases')

plt.legend(fontsize=12)
plt.show()
```

## 2c Plot Italy and US data on log axis

Note that we only want the y-axis to be logarithmic

```python
# your code here
# need plt.yscale('log')
plt.figure(figsize=(12,5))

plt.plot(data_italy.index,data_italy.values,'.-',color='C0',label='COVID-19 in Italy')
plt.plot(data_us.index,data_us.values,'.-',color='C1',label='COVID-19 in US')

plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('# of cases')
plt.yscale('log')

plt.legend(fontsize=12)
plt.show()
```

Use a for loop, and include x/y labels and a legend. Also save your figure as a jpg and share with friends. You are now a Python datascientist

```python
# set up df with 5 countries of interest
df_5 = df[df['Country/Region'].isin(['US','Italy','Germany','Turkey','Iran'])]
df_5 = df_5.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_5 = df_5.drop(columns=['Lat','Long','Province/State'])

data_5 = df_5.loc[:,'2/29/20':'4/14/20']
```

```python
data_5.head(5)
```

```python
data_5.iloc[2,:].name
```

```python
color5 = ['C0','C1','C2','C3','C4']

fig = plt.figure(figsize=(12,5))

for d in range(5):
    plt.plot(data_5.iloc[d,:].index,data_5.iloc[d,:].values,'.-',color=color5[d],label='COVID-19 in '+data_5.iloc[d,:].name)
    
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('# of cases')

plt.legend(fontsize=12)
plt.show()

# save your figure
fig.savefig('COVID-19_cases.jpg')
```
